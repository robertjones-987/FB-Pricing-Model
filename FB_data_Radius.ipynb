**Feature Selection**

```python
X = df.drop('Avg(clicks)', axis = 1).values
y = df['Avg(clicks)']
y.astype(int)
print(X.shape)
print(y.shape)
```

```python
location_dict = {}
for i, loc in enumerate(df["Location"].unique()):
    location_dict[loc] = i
```

```python
location_dict
```

```python
df["Location"] = df["Location"].map(location_dict)
```

```python
df[["Location", "Approx Reach(max)","Per day budget"]]
```

```python
# extract relevant columns
X = np.array(df[["Location", "Approx Reach(max)","Per day budget"]])
y = df['Avg(clicks)']
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
scale = std.fit_transform(X)
```

**Linear Model**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
model = LinearRegression()
```

```python
model
```

```python
model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = model.predict(X_test)
print(y_pred)
```

```python
X_test.shape
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred, multioutput = 'uniform_average', force_finite = True)
```

```python
print('Mean squared error: {:.2f}'.format(mse))
print('R2 score: {:.2f}'.format(r2))
```
